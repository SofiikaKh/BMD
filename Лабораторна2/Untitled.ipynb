```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
#Functional method
plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
```

```python
#multiply plots
plt.subplot(1,2,1)
plt.plot(x,y,"y")

plt.subplot(1,2,2)
plt.plot(y,x,"g")
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("y label")
axes.set_title("Title")
```

```python
fig=plt.figure()

axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.1,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_xlabel("X label")
axes1.set_ylabel("y label")
axes1.set_title("Title")

axes2.plot(y,x,"r")
axes2.set_xlabel("X label")
axes2.set_ylabel("y label")
axes2.set_title("Title")
```

```python
fig, axes = plt.subplots()
axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("y label")
axes.set_title("Title")
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,"r")
axes[1].plot(y,x)
```

```python
fig, axes=plt.subplots(nrows=4,ncols=4)
plt.tight_layout()
axes[0][1].plot(x,y,"r")
axes[2][2].plot(y,x,"r")
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label="x^2")
ax.plot(x,x**3, label="x^3")
ax.legend()
```

```python
fig=plt.figure(figsize=(8,4),dpi=100)
```

```python
fig, axes = plt.subplots()
ax.plot(x**2,color="b")
ax.plot(x**3,color="y")
```

```python
fig, ax = plt.subplots()
ax.plot(x,x**2,lw=2,ls="--",color="g",marker="o")
```

```python
fig, ax = plt.subplots()
ax.plot(x,x**2,color="g",marker="o",markersize=10,markerfacecolor="r")
fig.savefig("Plot_with_Marker")
```
