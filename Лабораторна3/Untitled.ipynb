```python
import seaborn as sns
%matplotlib inline
```

Завантаження даних з бібліотеки seaborn

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

Візуалізація даних
Графіки для аналізу кількісних даних:

```python
sns.distplot(tips["total_bill"])
```

Діаграма розсіювання:

```python
sns.jointplot(x="total_bill",y="tip",data=tips)
```

Попарні взаємозв’язки між змінними в повному наборі даних.
По діагоналі відображені графіки розподілу даних, в інших комірках- діаграми розсіювання (для відображення взаємозв’язку між двома змінними).

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue="sex")
```

Графіки для аналізу категоріальних даних:

Графік сукупних даних по кількісним даним залежності від категоріальної змінної.

```python
sns.barplot(x="sex",y="total_bill",data=tips)
```

```python
import numpy as np
sns.barplot(x="sex",y="total_bill",data=tips,estimator=np.median)
```

Графік розподілу категоріальних даних за певною кількісною змінною. Кожен «ящик» демонструє інтерквартильний розмах (25%, 50% та 75%), «вуса» демонструють повний розкид даних за виключенням викидів (аномально високі чи аномально низькі значення):

```python
sns.boxplot(x="day",y="total_bill",data=tips,palette="rainbow")
```

Діаграма розсіювання, де одна за змінних категоріальна:

```python
sns.stripplot(x="day",y="total_bill",data=tips)
```

Графіки для аналізу взаємозв’язку даних

Графік взаємозв’язку (кореляції) між даними у вигляді матриці.

```python
sns.heatmap(tips.corr())
```
